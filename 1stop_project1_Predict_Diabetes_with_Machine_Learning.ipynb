```python
#replace 0 values with the mean/median of the respective feature
#glucose
path["Glucose"]=path["Glucose"].replace(0,path["Glucose"].median())
#BloodPressure
path["BloodPressure"]=path["BloodPressure"].replace(0,path["BloodPressure"].median())
#BMI
path["BMI"]=path["BMI"].replace(0,path["BMI"].mean())
#SkinThickness
path["SkinThickness"]=path["SkinThickness"].replace(0,path["SkinThickness"].mean())
#Insulin
path["Insulin"]=path["Insulin"].replace(0,path["Insulin"].mean())
```

```python
path
```

```python
#splitting the dependent and independent variable
X=path.drop(["Outcome"],axis=1)
y=path["Outcome"]
```

```python
X
```

```python
y
```

```python
#splitting the data into training and testing paths
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
training_accuracy=[]
test_accuracy=[]
for n_neighbors in range(1,11):
  knn=KNeighborsClassifier(n_neighbors=n_neighbors)
  knn.fit(X_train,y_train)

  #check accuracy score
  training_accuracy.append(knn.score(X_train,y_train))
  test_accuracy.append(knn.score(X_test,y_test))
```

```python
plt.plot(range(1,11),training_accuracy,label="training_accuarcy")
plt.plot(range(1,11),test_accuracy,label="test_accuarcy")
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
print(knn.score(X_train,y_train),": training accuracy")
print(knn.score(X_test,y_test),": test accuracy")
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(random_state=0)
dt.fit(X_train,y_train)
print(dt.score(X_train,y_train),": training accuracy")
print(dt.score(X_test,y_test),": test accuracy")
```

```python
dt1= DecisionTreeClassifier(random_state=0,max_depth=3)
dt1.fit(X_train,y_train)
print(dt1.score(X_train,y_train),": training accuracy")
print(dt1.score(X_test,y_test),": test accuracy")
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(random_state=42)
mlp.fit(X_train,y_train)
print(mlp.score(X_train,y_train),": training accuracy")
print(mlp.score(X_test,y_test),": test accuracy")
```
